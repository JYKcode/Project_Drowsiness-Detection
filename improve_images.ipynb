# Image Preprocessing

- Histogram Normalizaion
- Bluring

```python
# Histogram equalization and bluring images with grayscale
import os
import sys
import cv2
```

```python
root_path = '/home/gyun941222/mildy/data'
data_path = os.path.join(root_path, 'e_data')

# hakerton 3class
train_path = os.path.join(data_path, 'train')
test_path = os.path.join(data_path, 'test')

# hakerton 4class improve
save_train_path = os.path.join(data_path, 'train_')
save_test_path = os.path.join(data_path, 'test_')
```

```python
for i in os.listdir(data_path):
    if i == 'train':
        for emotions in os.listdir(train_path):
            
            if emotions == 'train_ang': 
                emotion_path = os.path.join(train_path, emotions)

                for img in os.listdir(emotion_path): 
                    
                    img_path = os.path.join(emotion_path,img)
                    img_path = img_path.replace('\\', '/', 10)

                    img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) 
            
                    if img_array is None:
                        print('Image load failed!')
                        sys.exit()

                    dst_equal = cv2.equalizeHist(img_array)
                    dst = cv2.bilateralFilter(dst_equal, -1, 10, 5)
                    
                    save_path = os.path.join(save_train_path, emotions)
                    isExist = os.path.exists(save_path)
                    if not isExist:
                        os.makedirs(save_path)
                        
                    cv2.imwrite(f'{save_path}/{img}', dst)

            elif emotions == 'train_hap': 
                emotion_path = os.path.join(train_path, emotions)
                
                for img in os.listdir(emotion_path): 
                    
                    img_path = os.path.join(emotion_path,img)
                    img_path = img_path.replace('\\', '/', 10)

                    img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) 
            
                    if img_array is None:
                        print('Image load failed!')
                        sys.exit()

                    dst_equal = cv2.equalizeHist(img_array)
                    dst = cv2.bilateralFilter(dst_equal, -1, 10, 5)
                    
                    save_path = os.path.join(save_train_path, emotions)
                    isExist = os.path.exists(save_path)
                    if not isExist:
                        os.makedirs(save_path)
                        
                    cv2.imwrite(f'{save_path}/{img}', dst)
                    
            elif emotions == 'train_neu': 
                emotion_path = os.path.join(train_path, emotions)
                
                for img in os.listdir(emotion_path): 
                    
                    img_path = os.path.join(emotion_path,img)
                    img_path = img_path.replace('\\', '/', 10)

                    img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) 
            
                    if img_array is None:
                        print('Image load failed!')
                        sys.exit()

                    dst_equal = cv2.equalizeHist(img_array)
                    dst = cv2.bilateralFilter(dst_equal, -1, 10, 5)
                    
                    save_path = os.path.join(save_train_path, emotions)
                    isExist = os.path.exists(save_path)
                    if not isExist:
                        os.makedirs(save_path)
                        
                    cv2.imwrite(f'{save_path}/{img}', dst)    
                     
                   
    elif i == 'test':
        for emotions in os.listdir(test_path):
    
            if emotions == 'test_ang': 
                emotion_path = os.path.join(test_path, emotions)
                
                for img in os.listdir(emotion_path): 
                    
                    img_path = os.path.join(emotion_path,img)
                    img_path = img_path.replace('\\', '/', 10)

                    img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) 
            
                    if img_array is None:
                        print('Image load failed!')
                        sys.exit()

                    dst_equal = cv2.equalizeHist(img_array)
                    dst = cv2.bilateralFilter(dst_equal, -1, 10, 5)
                    
                    save_path = os.path.join(save_test_path, emotions)
                    isExist = os.path.exists(save_path)
                    if not isExist:
                        os.makedirs(save_path)
                        
                    cv2.imwrite(f'{save_path}/{img}', dst)

            elif emotions == 'test_hap': 
                emotion_path = os.path.join(test_path, emotions)
                
                for img in os.listdir(emotion_path): 
                    
                    img_path = os.path.join(emotion_path,img)
                    img_path = img_path.replace('\\', '/', 10)

                    img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) 
            
                    if img_array is None:
                        print('Image load failed!')
                        sys.exit()

                    dst_equal = cv2.equalizeHist(img_array)
                    dst = cv2.bilateralFilter(dst_equal, -1, 10, 5)
                    
                    save_path = os.path.join(save_test_path, emotions)
                    isExist = os.path.exists(save_path)
                    if not isExist:
                        os.makedirs(save_path)
                        
                    cv2.imwrite(f'{save_path}/{img}', dst)
                    
            elif emotions == 'test_neu': 
                emotion_path = os.path.join(test_path, emotions)
                
                for img in os.listdir(emotion_path): 
                    
                    img_path = os.path.join(emotion_path,img)
                    img_path = img_path.replace('\\', '/', 10)

                    img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) 
            
                    if img_array is None:
                        print('Image load failed!')
                        sys.exit()

                    dst_equal = cv2.equalizeHist(img_array)
                    dst = cv2.bilateralFilter(dst_equal, -1, 10, 5)
                    
                    save_path = os.path.join(save_test_path, emotions)
                    isExist = os.path.exists(save_path)
                    if not isExist:
                        os.makedirs(save_path)
                        
                    cv2.imwrite(f'{save_path}/{img}', dst)         
```
